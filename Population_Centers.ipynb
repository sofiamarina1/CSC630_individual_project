# Center by Human Population

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/sofia/2021-2022 School Year/CSC630/population_average_map/US_population.csv")
```

```python
df = df.replace(',','', regex=True)
```

```python
df['2019'] = pd.to_numeric(df['2019'])
```

```python
df
```

```python
pop_sum = 0
for i in df['Rank']:
    pop_sum = pop_sum + df['2019'][i-1]
pop_sum
```

```python
weights = []
for i in df['Rank']:
    weights.append(df['2019'][i-1] / pop_sum)
df['Weights'] = weights
```

```python
weighted_lats = 0
weighted_longs = 0
for i in df['Rank']:
    #print(df['Latitude'][i-1],df['Longitude'][i-1],df['2019'][i-1])
    weighted_lats = weighted_lats + (df['Latitude'][i-1] * df['Weights'][i-1])
    weighted_longs = weighted_longs + (df['Longitude'][i-1] * df['Weights'][i-1])
print('The center of the US weighted by population is: ({lat},{long})'.format(lat = weighted_lats, long = weighted_longs))
```
